Filling Missing Values in Features realted to Garage & Basement.

```python
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    data[col] = data[col].fillna(0)

for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    data[col] = data[col].fillna('None')

for col in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
    data[col] = data[col].fillna('None')
```

```python
data['MSZoning'] = data.groupby('MSSubClass')['MSZoning'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
from sklearn.impute import SimpleImputer # Import the SimpleImputer class from sklearn for handling missing values
cat_cols = ['Utilities','Exterior1st','Exterior2nd','MasVnrType','Electrical',
            'KitchenQual','Functional','SaleType']
# Initialize the SimpleImputer to replace missing values using the most frequent value of each column
imputer = SimpleImputer(strategy="most_frequent")

# Apply the imputer to the specified categorical columns of the dataset and update the data
data[cat_cols] = imputer.fit_transform(data[cat_cols])
```

Filling Missing Values in Numerical Columns.

```python
#Impute missing values  by the median of the same 'Neighborhood'

data["LotFrontage"] = data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

# Impute missing values for 'MasVnrArea' by the median of the same 'MasVnrType'
data["MasVnrArea"]  = data.groupby("MasVnrType")["MasVnrArea"].transform(lambda x: x.fillna(x.median()))

# Impute missing values for 'BsmtFinSF1' by the median of the same 'BsmtFinType1'
data["BsmtFinSF1"]  = data.groupby("BsmtFinType1")["BsmtFinSF1"].transform(lambda x: x.fillna(x.median()))

# Impute missing values for 'BsmtFinSF2' by the median of the same 'BsmtFinType2'
data["BsmtFinSF2"]  = data.groupby("BsmtFinType2")["BsmtFinSF2"].transform(lambda x: x.fillna(x.median()))
```

```python
# Fill missing values in 'BsmtFullBath' with 0.0, assuming no basement bathroom exists where data is missing
data["BsmtFullBath"] = data["BsmtFullBath"].fillna(0.0)

# Similarly, fill missing values in 'BsmtHalfBath' with 0.0, assuming no half bathrooms in the basement
data["BsmtHalfBath"] = data["BsmtHalfBath"].fillna(0.0)

# Calculate 'TotalBsmtSF' by adding 'BsmtFinSF1' and 'BsmtFinSF2', representing total finished basement area
data["TotalBsmtSF"]  = data["BsmtFinSF1"] + data["BsmtFinSF2"]

# Fill missing values in 'BsmtUnfSF' (unfinished basement area) with the median of the column
data["BsmtUnfSF"]    = data["BsmtUnfSF"].fillna(data["BsmtUnfSF"].median())
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

def evaluate_model(y_true, y_preds, X):
    mse = mean_squared_error(y_true, y_preds)  # Calculate Mean Squared Error
    mae = mean_absolute_error(y_true, y_preds)  # Calculate Mean Absolute Error
    rmse = np.sqrt(mse) # Calculate Root Mean Squared Error
    r2 = r2_score(y_true, y_preds)
    adjusted_r2 = 1 - (1-r2) * (len(y_true)-1) / (len(y_true) - X.shape[1] - 1)
    return r2, adjusted_r2, mae, mse, rmse

# Assuming X_test is your features test dataset for all model predictions
results = []

# Linear Regression Evaluation
linear_r2, linear_adjusted_r2, linear_mae, linear_mse, linear_rmse = evaluate_model(y_test, linear_predictions, X_test_encoded)
results.append({"Model": "Linear Regression", "R2": linear_r2, "Adjusted R2": linear_adjusted_r2, "MAE": linear_mae, "MSE": linear_mse, "RMSE": linear_rmse})

# Gradient Boosting Evaluation
gboost_r2, gboost_adjusted_r2, gboost_mae, gboost_mse, gboost_rmse = evaluate_model(y_test, gboost_predictions, X_test_encoded)
results.append({"Model": "Gradient Boosting", "R2": gboost_r2, "Adjusted R2": gboost_adjusted_r2, "MAE": gboost_mae, "MSE": gboost_mse, "RMSE": gboost_rmse})

# Random Forest Evaluation
rf_r2, rf_adjusted_r2, rf_mae, rf_mse, rf_rmse = evaluate_model(y_test, y_pred, X_test)
results.append({"Model": "Random Forest", "R2": rf_r2, "Adjusted R2": rf_adjusted_r2, "MAE": rf_mae, "MSE": rf_mse, "RMSE": rf_rmse})

# Create DataFrame
comparison_df = pd.DataFrame(results)
comparison_df.set_index("Model", inplace=True)
print(comparison_df)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import GridSearchCV

# Example of defining preprocessor and model
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X.select_dtypes(include=['number']).columns),
        ('cat', OneHotEncoder(handle_unknown='ignore'), X.select_dtypes(include=['object']).columns)
    ]
)# Scale numerical features to have mean 0 and variance 1
# One-hot encode categorical features, ignoring unknown categories during transform

# Initialize the Gradient Boosting Regressor with a fixed random state
gboost_model = GradientBoostingRegressor(random_state=42)

model_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', gboost_model)
])


# Set up a grid of hyperparameters to test during Grid Search
param_grid = {
    'model__n_estimators': [100, 500, 1000],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__max_depth': [3, 4, 5],
    'model__min_samples_leaf': [3, 5, 9],
    'model__min_samples_split': [2, 6, 10]
}


 #Configure Grid Search with the model pipeline, parameter grid, 3-fold cross-validation,
# and scoring based on negative Mean Squared Error
grid_search = GridSearchCV(model_pipeline, param_grid, cv=3, scoring='neg_mean_squared_error', verbose=1)
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score (negative MSE):", grid_search.best_score_)
```
